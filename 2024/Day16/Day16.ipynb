```python
r_coord, r_direct = len(lines)-2+1j,1j
tgt_coord = 1+(len(lines[0])-2)*1j
area.get_value(r_coord),area.get_value(tgt_coord)
```

```python
def score_one(p1,p2):
    coord_1,direct_1 = p1
    coord_2,direct_2 = p2
    #return abs(coord_1.real-coord_2.real)+abs(coord_1.imag-coord_2.imag)+int(direct_1 != direct_2)*1000
    return abs(coord_1-coord_2)+int(direct_1 != direct_2)*1000
```

```python
edges = set()
edges_no_weight = set()
for coord,val in area.map_as_set():
    if val == "#": continue

    for direct in [1,-1,1j,-1j]:
        n_next = set((coord+d,d) for d in [direct * mult for mult in [1j,1,-1j]])
        n_next = set((c,d) for (c,d) in n_next if area.get_value(c)!= "#")

        p1 = (coord,direct)
        for n_coord,n_d in n_next:
            p2=(n_coord,n_d)
            if (p2,p1) not in edges_no_weight:
                edges.add((p1,p2,score_one(p1,p2)))
                edges_no_weight.add((p1,p2))
                edges_no_weight.add((p2,p1))

print(len(edges))             
            
```

```python
G=nx.Graph()
for p1,p2,w in edges:
    G.add_edge(p1,p2,weight=w)
```

```python
pth_1 = nx.shortest_path(G,source=(r_coord,r_direct),target=(tgt_coord,1j),weight="weight")
score_1 = nx.path_weight(G,pth_1,weight="weight")

pth_2 = nx.shortest_path(G,source=(r_coord,r_direct),target=(tgt_coord,-1),weight="weight")
score_2 = nx.path_weight(G,pth_2,weight="weight")

if score_1 < score_2:
    pth,score = pth_1,score_1
else:
    pth,score = pth_2,score_2

score
```
