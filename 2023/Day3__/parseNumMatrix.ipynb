```python
matr =[]
with open('input.txt') as file:
    for line in file:
        line = line.replace("\n","")
        matr.append(list(line))
print(matr)
```

go row by row and gather numbers

```python
def checkSurrounded(matr_in,i_row,i_col,num_rows,num_cols):
    checks_dr = [(i_row+x,i_col+y) for x in [-1,0,1] for y in [-1,0,1]]
    checks_conf = [(x,y) for (x,y) in checks_dr if x >= 0 and y >= 0 and x < num_rows and y < num_cols]
    checks_conf.remove((i_row,i_col))
    #print(checks_dr)
    #print(checks_conf)
    
    for x,y in checks_conf:
        c = matr_in[x][y]
        if c == "." or c.isdigit():
            continue
        else:
            return True
    return False

checkSurrounded(matr,0,2,10,9)
#checkSurrounded(matr,9,9,10,9)
```

```python
def parseRow(matr_in,i_row, i_col_in_row): 
    my_num = "0"
    is_surround = False
    num_cols=len(matr_in[i_row])
    num_rows = len(matr_in)
    
    while i_col_in_row < num_cols :
        line = matr_in[i_row][i_col_in_row:]
        #print(i_col_in_row,line)
        if line[0].isdigit():
            my_num += line[0]
            is_surround = (is_surround or checkSurrounded(matr_in,i_row,i_col_in_row,num_rows,num_cols))
            i_col_in_row += 1
            #print(">>",my_num,is_surround)
            
        elif is_surround:
            #print("<<",my_num)
            return int(my_num) + parseRow(matr_in,i_row,i_col_in_row+1)
        else:
            return parseRow(matr_in,i_row,i_col_in_row+1)
    
    # if number was last thing in the  line
    return (int(my_num) if is_surround else 0)
    
parseRow(matr,2,5)
```

```python
my_sum=0
for i,_ in enumerate(matr):    
    my_sum += parseRow(matr,i,0)
my_sum
```

```python
def findNumber(matr_in,i_row,i_col):
    
    for j in range(i_col,-1,-1):
        if not matr_in[i_row][j].isdigit():
            j+=1
            break
    #print(i_col)
    #print(j)
        
    for k in range(i_col,len(matr_in[0]),1):
        if not matr_in[i_row][k].isdigit():
            k -= 1
            break
    #print(k)
    #print(matr_in[i_row][j:k+1])
    
    return int("".join(matr_in[i_row][j:k+1]))

    #return matr_in[i_row:i_row+1][0][i_col:i_col+1]
#findNumber(matr,0,1)
```

```python
def adjTo2Nums(matr_in,i_row,i_col):
    num_rows = len(matr_in)
    num_cols = len(matr_in[0])
    
    checks_dr = [(i_row+x,i_col+y) for x in [-1,0,1] for y in [-1,0,1]]
    checks_conf = [(x,y) for (x,y) in checks_dr if x >= 0 and y >= 0 and x < num_rows and y < num_cols]
    checks_conf.remove((i_row,i_col))
    
    checks_num = [(x,y) for (x,y) in checks_conf if matr_in[x][y].isdigit()]
    
    checks_final = checks_num.copy()
    
    for x,y in checks_dr:
        if (x,y) in checks_final and (x,y+1) in checks_final:
            checks_final.remove((x,y))
    if len(checks_final) != 2:
        return 0
    else:
        x_1,y_1 = checks_final[0]
        x_2,y_2 = checks_final[1]
        return findNumber(matr_in,x_1,y_1)*findNumber(matr_in,x_2,y_2)

adjTo2Nums(matr,1,3)
```
