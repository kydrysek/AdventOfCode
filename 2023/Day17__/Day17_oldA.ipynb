```python
def buildArrayFromLines(lines):
    
    tmp_arr = []
    for line in lines:
        tmp_line = []    
        
        for c in line:
            if c != "\n":
                tmp_line += [int(c)]
        tmp_arr.append(tmp_line)
    
    
    return np.array(tmp_arr)
    
```

```python
# OK
def parseFile(fileName):   
    with open(fileName) as file:
        lines =  file.readlines()    
   
    my_arr = pd.DataFrame(buildArrayFromLines(lines))
    
    
    return my_arr
```
