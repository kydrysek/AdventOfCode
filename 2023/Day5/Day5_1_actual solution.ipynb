```python
import re

#FILENAME="test.txt"
FILENAME="input.txt"
```

## Version 1

```python
state=0 #-1 seeds 0-empty line 1-header 2-numbers

with open(FILENAME) as file:
    #1. get seeds
    searched = re.search("^seeds:(.*)$",file.readline())
    seed_list = [int(x) for x in searched.group(1).split()]
    
    result_dict = dict(zip(seed_list,seed_list))
    #print(result_dict)
    state=-1
    seed_tobe_mapped = {}
    for ll in file:
        if ll.strip()=="":
            state = 0
        else:
            if state not in [0,1,2]:
                pass
            elif state==0:
                #starting next group
                print("<<",result_dict,"\n")
                print(ll[:-1])
                searched = re.search("^(\w+)-to-(\w+) map:$",ll)
                #print(searched.group(1),"--->",searched.group(2),"..")
                state=1
                seed_tobe_mapped = dict.fromkeys(seed_list,1)
                #print(seed_tobe_mapped)
            else:
                #print(ll)
                searched = re.search("^(\d+) (\d+) (\d+)$",ll)
                tgt_first = int(searched.group(1))
                src_first = int(searched.group(2))
                item_count = int(searched.group(3))
                
                #print("<<",result_dict)
                #print("<<",seed_tobe_mapped)
                for seed_id,seed_map in result_dict.items():
                    if seed_tobe_mapped[seed_id]:
                        x = result_dict[seed_id] 
                        #print(tgt_first,src_first,item_count,"x=",x)
                        if x >= src_first and (x < src_first + item_count):
                            seed_tobe_mapped[seed_id] = 0
                            result_dict[seed_id] = tgt_first + (x-src_first)
                    
                
                state=2
        
#print(scratch_count.values())
val = min(result_dict.values())
idx = max([i for i in seed_list if result_dict[i]==val])
print(idx,val)
```
