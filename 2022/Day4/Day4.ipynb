```python
import re
from typing import Set, List, Tuple, Match
from operator import itemgetter

# FILENAME : str ="test.txt"
FILENAME : str ="input.txt"
DIAG : bool =False
```

```python
def processFile(file_name : str) -> List[str]:
    with open(file_name) as file:
        lines = file.readlines()
    
    lines_better = map(lambda x:x.replace("\n",""),lines)
        
    return list(lines_better)
    
```

```python
def getRangePairs(lines : List[str]) -> List[Tuple[Tuple[int,int],...]]:
    pattern : str = "^(\d+)-(\d+),(\d+)-(\d+)$"
    
    res : List[Tuple[Tuple[int,int],...]] = []
    for line in lines:
        mmm = re.search(pattern,line)
        
        if mmm is not None: 
            mmm_i = list(map(int,mmm.groups()))
            res.append(((mmm_i[0],mmm_i[1]),(mmm_i[2],mmm_i[3])))
    
    return res
    
```

```python
getRangePairs(["2-8,3-7"])
```

```python
def rangeContainsAnother(range1:Tuple[int,int],range2:Tuple[int,int]) -> bool:
    if range1[0]>=range2[0] and range1[1] <= range2[1]:
        return True
    elif range2[0]>=range1[0] and range2[1] <= range1[1]:
        return True
    else:
        return False
    
def rangeContainsAnother_tupoftup(ranges) -> bool:
    return rangeContainsAnother(ranges[0],ranges[1])
```

```python
def rangesOverlap(range1:Tuple[int,int],range2:Tuple[int,int]) -> bool:
    if range1[1]<range2[0] or range1[0]>range2[1]:
        return False
    else:
        return True
    
def rangesOverlap_tupoftup(ranges) -> bool:
    res =  rangesOverlap(ranges[0],ranges[1])
    if DIAG: print(ranges[0],ranges[1],res)
    return res
```

```python
lines = processFile(FILENAME)
rangePairs = getRangePairs(lines)
rangePairs
# print(rangePairs)
resA = sum(map(rangeContainsAnother_tupoftup,rangePairs))
print("Result partA:",resA)

resB = sum(map(rangesOverlap_tupoftup,rangePairs))
print("Result partB:",resB)
```
